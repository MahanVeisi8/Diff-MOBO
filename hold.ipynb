```python
file_results = np.load("run_results_xstrain_final.npy",allow_pickle=True).item()
xs_train = np.load(rf"data/xs_train.npy",allow_pickle=True)
```

```python
file_shapes = file_results["shapes"]
file_performances = file_results["performances"]
```

```python
for i in range(len(file_shapes)):
    shape = file_shapes[i]
    performances = file_performances[i]
    real_shape = xs_train[i]
    if not np.array_equal(shape, real_shape):
        print(f"hi {i=}")
        break
```

```python
indices =[]
for i in range(len(file_performances)):
    if file_performances[i][0]== -1000:
        continue
    else:
        indices.append(i)
```

```python
len(indices) ,len(xs_train)
```

```python
valid_shapes = file_shapes[indices]
valid_performances = file_performances[indices][:,:2]
valid_shapes.shape , valid_performances.shape 
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Assuming you already have:
# valid_shapes = np.load('valid_shapes.npy')
# valid_performances = np.load('valid_performances.npy')

# Choose test size (e.g., 3000 samples)
# test_size = 3000 / len(valid_shapes)  # proportional split

# Split the data
shapes_train, shapes_test, perf_train, perf_test = train_test_split(
    valid_shapes,
    valid_performances,
    test_size=0.1,
    random_state=42,
    shuffle=True
)

# Optionally save the splits
np.save('xs_train_MPI.npy', shapes_train)
np.save('xs_test_MPI.npy', shapes_test)
np.save('ys_train_MPI.npy', perf_train)
np.save('ys_test_MPI.npy', perf_test)

print(f"Train set: {len(shapes_train)} samples")
print(f"Test set: {len(shapes_test)} samples")
```
